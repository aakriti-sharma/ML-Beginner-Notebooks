```python
import numpy as np  #importing libraries
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import io 
```

```python
img = cv2.imread('./Downloads/pizza.jpg')    #reading source image
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   
io.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)   
gray = cv2.medianBlur(gray, 5)
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)  #detecting the edges
io.imshow(edges)
```

```python
color = cv2.bilateralFilter(img, 9, 250, 250)
cartoon = cv2.bitwise_and(color, color, mask=edges)      #overlapping edges with bilateral filter transformed image
```

```python
io.imshow(cartoon)    #output animated image
cv2.imwrite("piz-cartoon.jpg",cartoon)
```
